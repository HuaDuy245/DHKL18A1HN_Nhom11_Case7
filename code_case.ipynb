```python
import numpy as np
```

```python
# 1. Đọc dữ liệu từ hai file
print("\n[Nhiệm vụ 1] Đọc dữ liệu từ file CSV...")
du_lieu_fb = np.genfromtxt('facebook_sentiment.csv', delimiter=',', skip_header=1, dtype=None, encoding=None)
du_lieu_tw = np.genfromtxt('twitter_sentiment.csv', delimiter=',', skip_header=1, dtype=None, encoding=None)
```

```python
# 2. Tách time và sentiment
print("[Nhiệm vụ 2] Tách mảng thời gian và sentiment.")
thoi_gian_fb_chuoi = du_lieu_fb['f0']
cam_xuc_fb = du_lieu_fb['f1']
thoi_gian_tw_chuoi = du_lieu_tw['f0']
cam_xuc_tw = du_lieu_tw['f1']

```

```python
# 3. Chuyển time sang datetime64
print("[Nhiệm vụ 3] Chuyển đổi Dtype cột Time sang datetime64.")
thoi_gian_fb = thoi_gian_fb_chuoi.astype('datetime64[m]') 
thoi_gian_tw = thoi_gian_tw_chuoi.astype('datetime64[m]')
```

```python
# 4. Kiểm tra dtype, shape; in 5 giá trị đầu cuối.
print("\n[Nhiệm vụ 4] Kiểm tra Dtype, Shape và In 5 giá trị Đầu/Cuối.")

def in_thong_tin_mang(cam_xuc, nen_tang):
    print(f"\n--- Thông tin Array {nen_tang} ---")
    print(f"  Dtype Sentiment: {cam_xuc.dtype}, Shape: {cam_xuc.shape}")
    print(f"  5 giá trị đầu (Sentiment): {cam_xuc[:5]}")
    print(f"  5 giá trị cuối (Sentiment): {cam_xuc[-5:]}")

in_thong_tin_mang(cam_xuc_fb, "Facebook")
in_thong_tin_mang(cam_xuc_tw, "Twitter")

```

```python
# 5. Slicing 20% đầu – 20% giữa – 20% cuối.
print("\n[Nhiệm vụ 5] Slicing 20% Đầu, 20% Giữa, 20% Cuối.")

def cat_cam_xuc(mang_cam_xuc, ten_nen_tang):
    N = len(mang_cam_xuc)
    phan_tram_20 = int(N * 0.2)
    bat_dau_giua = N // 2 - phan_tram_20 // 2
    ket_thuc_giua = bat_dau_giua + phan_tram_20

    print(f"--- {ten_nen_tang} (20% = {phan_tram_20} bản ghi) ---")
    print(f"  20% Đầu (5 giá trị): {mang_cam_xuc[:phan_tram_20][:5]}")
    print(f"  20% Giữa (5 giá trị): {mang_cam_xuc[bat_dau_giua:ket_thuc_giua][:5]}")
    print(f"  20% Cuối (5 giá trị): {mang_cam_xuc[-phan_tram_20:][:5]}")

cat_cam_xuc(cam_xuc_fb, "Facebook")
cat_cam_xuc(cam_xuc_tw, "Twitter")

```

```python
# 6. Boolean Masking: Rất Tích Cực (> 0.8) và Rất Tiêu Cực (< -0.8).
print("\n[Nhiệm vụ 6] Boolean Masking: Rất Tích Cực (> 0.8) và Rất Tiêu Cực (< -0.8).")

dem_fb_tich_cuc = np.sum(cam_xuc_fb > 0.8)
dem_fb_tieu_cuc = np.sum(cam_xuc_fb < -0.8)
dem_tw_tich_cuc = np.sum(cam_xuc_tw > 0.8)
dem_tw_tieu_cuc = np.sum(cam_xuc_tw < -0.8)

print(f"  Facebook: Rất Tích Cực ({dem_fb_tich_cuc}), Rất Tiêu Cực ({dem_fb_tieu_cuc})")
print(f"  Twitter: Rất Tích Cực ({dem_tw_tich_cuc}), Rất Tiêu Cực ({dem_tw_tieu_cuc})")

```

```python
# 7. Tạo bảng tổng hợp sentiment và phân loại bình luận theo mức độ.
print("\n[Nhiệm vụ 7] Phân loại Sentiment theo 5 Mức độ.")

def phan_loai_cam_xuc(mang_cam_xuc, ten_nen_tang):
    r_tc = np.sum(mang_cam_xuc > 0.6)
    tc = np.sum((mang_cam_xuc > 0.2) & (mang_cam_xuc <= 0.6))
    trung_tinh = np.sum((mang_cam_xuc >= -0.2) & (mang_cam_xuc <= 0.2))
    tieu_cuc = np.sum((mang_cam_xuc < -0.2) & (mang_cam_xuc >= -0.6))
    r_tieu_cuc = np.sum(mang_cam_xuc < -0.6)
    total = len(mang_cam_xuc)

    ket_qua_phan_loai = np.array([
        ['Rất Tích Cực (> 0.6)', r_tc, r_tc / total * 100],
        ['Tích Cực (0.2 -> 0.6)', tc, tc / total * 100],
        ['Trung Tính (-0.2 -> 0.2)', trung_tinh, trung_tinh / total * 100],
        ['Tiêu Cực (-0.6 -> -0.2)', tieu_cuc, tieu_cuc / total * 100],
        ['Rất Tiêu Cực (<= -0.6)', r_tieu_cuc, r_tieu_cuc / total * 100]
    ], dtype=object)

    print(f"\n--- Phân loại Sentiment {ten_nen_tang} (Tổng: {total}) ---")
    header = ["Mức độ", "Số lượng", "Tỷ lệ (%)"]
    print("{:<30} {:<10} {:<10}".format(*header))
    print("-" * 50)
    for dong in ket_qua_phan_loai:
        print("{:<30} {:<10} {:<10.2f}".format(dong[0], dong[1], dong[2]))

phan_loai_cam_xuc(cam_xuc_fb, "Facebook")
phan_loai_cam_xuc(cam_xuc_tw, "Twitter")

```

```python
# 8. Fancy indexing lấy top-20 bình luận tích cực và tiêu cực nhất.
print("\n[Nhiệm vụ 8] Fancy Indexing: Top-20 Tích Cực và Tiêu Cực nhất.")

def lay_top_20(mang_cam_xuc, ten_nen_tang):
    chi_so_sap_xep = np.argsort(mang_cam_xuc)
    
    cam_xuc_tieu_cuc_nhat = mang_cam_xuc[chi_so_sap_xep[:20]]
    cam_xuc_tich_cuc_nhat = mang_cam_xuc[chi_so_sap_xep[-20:][::-1]]

    print(f"\n--- {ten_nen_tang} ---")
    print(f"  Top 5 Tích Cực nhất: {cam_xuc_tich_cuc_nhat[:5]}")
    print(f"  Top 5 Tiêu Cực nhất: {cam_xuc_tieu_cuc_nhat[:5]}")

lay_top_20(cam_xuc_fb, "Facebook")
lay_top_20(cam_xuc_tw, "Twitter")

```

```python
# 9. np.diff phân tích biến động | Δsentiment| > 0.6.
print("\n[Nhiệm vụ 9] np.diff phân tích biến động |ΔSentiment| > 0.6.")

def phan_tich_bien_dong(mang_cam_xuc, ten_nen_tang):
    bien_dong = np.diff(mang_cam_xuc)
    mask_bien_dong_lon = np.abs(bien_dong) > 0.6
    
    dem_bien_dong_lon = np.sum(mask_bien_dong_lon)
    print(f"  {ten_nen_tang}: Số lượng biến động |Δs| > 0.6: {dem_bien_dong_lon} lần.")
    
phan_tich_bien_dong(cam_xuc_fb, "Facebook")
phan_tich_bien_dong(cam_xuc_tw, "Twitter")


```

```python
# 10. Thống kê tổng hợp: mean, std, median, min, max.
print("\n[Nhiệm vụ 10] Tính toán Thống kê Tổng hợp Sentiment.")

def tinh_thong_ke(mang_cam_xuc, ten_nen_tang):
    trung_binh = np.mean(mang_cam_xuc)
    do_lech_chuan = np.std(mang_cam_xuc)
    trung_vi = np.median(mang_cam_xuc)
    toi_thieu = np.min(mang_cam_xuc)
    toi_da = np.max(mang_cam_xuc)

    print(f"\n--- Thống kê {ten_nen_tang} ---")
    print(f"  Trung bình (Mean): {trung_binh:.4f}")
    print(f"  Độ lệch chuẩn (Std): {do_lech_chuan:.4f}")
    
    return np.array([trung_binh, do_lech_chuan, trung_vi, toi_thieu, toi_da])

thong_ke_fb = tinh_thong_ke(cam_xuc_fb, "Facebook")
thong_ke_tw = tinh_thong_ke(cam_xuc_tw, "Twitter")


```

```python
# 11. So sánh sentiment 2 nền tảng.
print("\n[Nhiệm vụ 11] So sánh Sentiment giữa Facebook và Twitter.")

nhan_thong_ke = np.array(["Mean", "Std", "Median", "Min", "Max"])
du_lieu_so_sanh = np.vstack((thong_ke_fb, thong_ke_tw))

print("\n--- Bảng So Sánh Thống Kê Sentiment ---")
print("{:<10} {:<10} {:<10}".format("Chỉ số", "Facebook", "Twitter"))
print("-" * 30)
for i in range(len(nhan_thong_ke)):
    print("{:<10} {:<10.4f} {:<10.4f}".format(nhan_thong_ke[i], du_lieu_so_sanh[0, i], du_lieu_so_sanh[1, i]))


```

```python
print("\n[Nhiệm vụ 12] Xuất file kết quả: ket_qua_case_7.csv")

mang_ket_qua = np.array([
    ["Chỉ số", "Facebook", "Twitter"],
    ["Mean", thong_ke_fb[0], thong_ke_tw[0]],
    ["Std", thong_ke_fb[1], thong_ke_tw[1]],
    ["Median", thong_ke_fb[2], thong_ke_tw[2]],
    ["Min", thong_ke_fb[3], thong_ke_tw[3]],
    ["Max", thong_ke_fb[4], thong_ke_tw[4]],
], dtype=object)

with open("ket_qua_case_7.csv", "w", encoding="utf-8") as f:
    np.savetxt(f, mang_ket_qua, fmt='%s', delimiter=',')
```
